```python
import xgboost as xgb
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
import pickle
```

```python
# create xgboost classifier object
xgb_clf=xgb.XGBClassifier()
```

```python
param_grid={'max_depth':[3,5,7],'min_child_weight':[1,2,5],'subsample':[0.5,0.7],'learning_rate':[0.01,0.1,0.5]}
```

```python
grid_cv=GridSearchCV(xgb_clf,param_grid,scoring='accuracy',cv=5,n_jobs=-1)
grid_cv.fit(x_train,y_train)
print(grid_cv.best_params_)
final_cl=xgb.XGBClassifier(**grid_cv.best_params_)
final_cl.fit(x_train,y_train)
y_pred=final_cl.predict(x_test)
acc_scr=accuracy_score(y_test,y_pred)
print(acc_scr)
```

```python
filename='Global_power_plant_linaer.pkl'
pickle.dump(final_cl,open(filename,'wb'))
load_model=pickle.load(open(filename,'rb'))
predictions=load_model.predict(x_test)
print(predictions)
```

we can see that the XGBooster is giving the good accuarcy
