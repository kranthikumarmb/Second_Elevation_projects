```python
import warnings
warnings.filterwarnings('ignore')
df.fillna(df.mean(),inplace=True)
```

```python
df['geolocation_source'].fillna(df['geolocation_source'].mode()[0],inplace=True)
df['generation_data_source'].fillna(df['generation_data_source'].mode()[0],inplace=True)
```

```python
from numpy import percentile
columns=df.columns
for j in columns:
    if isinstance(df[j][0],str):
        continue
        
    else:
        qua=percentile(df[j],[25,75]) 
        low=qua[0]-(1.5*qua[1]-qua[0])
        upper=qua[0]-(1.5*qua[1]-qua[0])
        df[j]=df[j].apply(lambda x:upper if x>upper else (low if x<low else x))
```

```python
from sklearn import preprocessing
```

```python
le=preprocessing.LabelEncoder()
```

```python


for col in df.columns:
    df[col]=le.fit_transform(df[col])
```

```python
x=df.drop('capacity_mw',axis=1).copy()
y=df['capacity_mw'].copy()
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
y_test=np.array(y_test,dtype=float)
```

```python
import sklearn.metrics as metrics 
```

```python
def regression_result(y_true,y_pred):
    explained=metrics.explained_variance_score(y_true,y_pred)
    mean_absolute=metrics.mean_absolute_error(y_true,y_pred)
    mse=metrics.mean_squared_error(y_true,y_pred)
    r2=metrics.r2_score(y_true,y_pred)
    adjusted=1-(1-r2)*(len(y_true)-1)/(len(y_true)-x_test.shape[1]-1)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
def model_accuracy(model,x_train=x_train,y_train=y_train):
    acc=cross_val_score(estimator=model,x=x_train,y=y_train,cv=10)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
import statsmodels.api  as sm
```

```python
reg=LinearRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
from yellowbrick.regressor import ResidualsPlot
```

```python
fig=plt.figure(figsize=(6,5))
v=ResidualsPlot(reg,hist=True,qqplot=False)
v.score(x_test,y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rc=RandomForestRegressor()
rc.fit(x_train,y_train)
```

```python
fig=plt.figure(figsize=(12,7))
vs=ResidualsPlot(rc,hist=True,qqplot=False)
vs.score(x_test,y_test)
vs.show()
```

# Decision tree 

```python
from sklearn.tree import DecisionTreeRegressor
dc=DecisionTreeRegressor()
dc.fit(x_train,y_train)
```

```python
fig=plt.figure(figsize=(8,6))
vsc=ResidualsPlot(dc,hist=True,qqplot=False)
vsc.score(x_test,y_test)
vsc.show()
```

```python
import pickle
```

```python
filename='GlobalPowerPlant.pkl'
pickle.dump(dc,open(filename,'wb'))
```
